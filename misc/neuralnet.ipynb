```python
# Sigmoid (or logistic) neuron
class N(object):
    def __init__(self, b, *w):
        self.bias = b
        self.weights = w
    def __call__(self, *x):
        return 1./(1 + exp(-dot(self.weights,x)-self.bias))
```

```python
# "soft" NAND
NAND = N(3, -2, -2)

xx = arange(0,1,.1)
w = xx.shape[0]

grid = zeros((w, w))

for i in range(w):
    for j in range(w):
        grid[w-i-1,j] = NAND(xx[i],xx[j])

imshow(grid, extent=(0,1,0,1))

colorbar()
```
